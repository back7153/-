##### 슬라이스
- s[i:j] : s[i]부터 s[j-1]까지 나열
- s[i:j:k] : s[i]부터 s[j-1]까지 k씩 건너뛰며 나열
- i, j가 len(s)보다 크면 len(s)가 지정된 것으로 간주
- i가 없거나 None이면 0이 지정된 것
- j가 없거나 None이면 len(s)가 지정된 것
- s[:] : s의 모든 원소를 출력
- s[-n:] : s의 원소 중 뒤에서 n번째 원소부터 맨 뒤의 원소까지 출력
- s[::-1] : s의 모든 원소를 거꾸로 출력

- 배열 원소의 최댓값 구하기
```python
# a의 원소가 3개일 때 최댓값 구하기
maximum = a[0]
if a[1] > maximum:
    maximum = a[1]
if a[2] > maximum:
    maximum = a[2]
```
n개의 원소가 있다면 if문을 n-1번 실행해야 한다.

````python
# 배열 a의 원소들 중 최댓값을 구하는 함수
def max_of(a):
    maximum = a[0]
    for i in range(1, len(a)):
        if a[i] > maximum:
            maximum = a[i]
    return maximum

if __name__ == '__main__':
    print('배열의 최댓값을 구합니다.')
    num = int(input('원소 수를 입력하세요.'))
    x = [None] * num
    
    for i in range(num):
        x[i] = int(input(f'x[{i}]값을 입력하세요.: '))

    print(f'최댓값은 {max_of(x)}입니다.')
````

- 배열 원소를 역순으로 정렬하기

    ex) [2, 5, 1, 3, 9, 6, 7] -> [7, 6, 9, 3, 1, 5, 2]

````python
def reverse_array(a):
    n = len(a)
    for i in range(n // 2):
        a[i], a[n - 1 - i] = a[n - 1 - i], a[i]

if __name__ == '__main__':
    print('배열 원소를 역순으로 정렬합니다.')
    nx = int(input('원소 수를 입력하세요.: '))
    x = [None] * nx

    for i in range(nx):
        x[i] = int(input(f'x[{i}]의 값을 입력하세요.: '))

    reverse_array(x)

    print('배열의 원소를 역순으로 정렬했습니다.')
    for i in range(nx):
        print(f'x[{i}] = {x[i]}')
````

- 기수 변환하기(n진수 구하기)

````python
# 10진수 정숫값을 입력받아 2~36진수로 변환하여 출력하기

def card_conv(x: int, r: int) -> str:
    '''
    정숫값 x를 r진수로 변환한 뒤 그 수를 문자열 타입으로 반환
    '''

    d = '' # 빈 문자열 생성
    dchar = '0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZ' # 0 ~ 9 : 10개 + A ~ Z : 26 = 36개

    while x > 0: # x가 0보다 큰 경우에는 계속 반복
        d += dchar[x % r] # x 나누기 r의 나머지를 인덱스로 해서 dchar의 문자열 중 하나의 문자열을 d 문자열에 추가
        x //= r # x 나누기 r의 몫을 x에 할당해서 다시 반복
    
    return d[::-1] # 반환은 역순으로

if __name__ == '__main__':
    print('10진수를 n진수로 변환합니다.')

    while True: # 계속 반복
        while True: # 음수를 입력한 경우 계속 반복
            no = int(input("음이 아닌 정수를 입력하세요.: ")) 
            if no > 0: 
                break # 음이 아닌 정수가 입력되면 while문 탈출
        
        while True: # 2 ~ 36이 아닌 숫자가 입력됐을 경우 계속 반복
            cd = int(input("어떤 진수로 변환할까요?: "))
            if 2 <= cd <= 36:
                break # 2 ~ 36 사이의 값이 입력되면 while문 탈출
        
        print(f'{cd}진수로는 {card_conv(no, cd)}입니다.')

        retry = input("한 번 더 변환할까요?(Y : 예 / N : 아니요): ")

        if retry in {'N', 'n'}: 
            break # N, n이 입력됐을 경우 while문 탈출
````

````python
# 진수 변환 과정을 보여주는 함수

def card_conv_verbose(x: int, r: int) -> str:
    '''
    정숫값 x를 r진수로 변환한 뒤 그 수를 문자열 타입으로 반환
    '''

    d = '' # 빈 문자열 생성
    dchar = '0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZ' # 0 ~ 9 : 10개 + A ~ Z : 26 = 36개
    n = len(str(x))

    print(f'{r:2} | {x:{n}d}')

    while x > 0: # x가 0보다 큰 경우에는 계속 반복
        print('   +' + (n + 2) * '-')
        if x //r :
            print(f'{r:2} | {x // r:{n}d} ... {x % r}')
        else:
            print(f'     {x // r:{n}d} ... {x % r}')
        d += dchar[x % r] # x 나누기 r의 나머지를 인덱스로 해서 dchar의 문자열 중 하나의 문자열을 d 문자열에 추가
        x //= r # x 나누기 r의 몫을 x에 할당해서 다시 반복
    
    return d[::-1] # 반환은 역순으로

card_conv_verbose(29, 16)
````

- 소수(Prime Number) 나열하기

    - 소수의 조건 : **어떤 정수 n이 소수라면 n은 2부터 n-1까지의 어떤 정수로도 나누어 떨어지지 않는다.**
    - 합성수(Composite Number) : 소수가 아닌 수

````python
# 1000 이하의 소수 나열하기

counter = 0 # 나누기를 수행한 횟수

for n in range(2, 1001): # 2 ~ 1000까지 숫자(1은 소수가 아님)
    for i in range(2, n): # 2부터 2, 2부터 3, 2부터 4, ..., 2부터 1000까지
        counter += 1 # 나누기 횟수 +1
        if n % i == 0: # n을 i로 나누었을 때 나머지가 0이면(= 나누어 떨어지면)
            break # 내부 for문을 탈출(한 번이라도 나누어 떨어지면 소수가 아니므로 탈출)
    else: # 내부 for문을 중간에 탈출한게 아니라면
        print(n) # n을 출력(소수이므로)

print(f'나눗셈을 실행한 횟수: {counter}')
````

- 소수 찾기 알고리즘 개선 1
    - 개선점 : **n이 2와 3으로 나누어 떨어지지 않는다면 $2 \times 2 = 4$와 $2 \times 3 = 6$으로도 나누어 떨어지지 않는다.**
    - 소수의 조건 : **어떤 정수 n이 소수라면 n은 2부터 n-1까지 <u>어떤 소수</u>로도 나누어 떨어지지 않는다.**
    - ex) 7이 소수인지 판단하려면?
        - 7보다 작은 소수인 2, 3, 5로 나눗셈을 해서 나누어 떨어지지 않는 것을 보이면 된다.
        - $( \because )$ 2와 3으로 나누어 떨어지지 않으므로 4와 6은 해보지 않아도 된다.

````python
# 1000 이하의 소수를 나열하기(알고리즘 개선 1)

counter = 0 # 나누기를 수행한 횟수
ptr = 0 # 먼저 찾은 소수의 개수
prime = [None] * 500 # prime : 소수를 저장할 리스트

prime[ptr] = 2 # prime 리스트의 첫번째 원소는 2
ptr += 1 # 2가 있으므로 찾은 소수의 개수는 1

for n in range(3, 1001, 2): # 3, 5, 7, 9, 11, 13, ... , 999(= 홀수만)
    for i in range(1, ptr): # 1부터 (찾은 소수의 개수-1)까지
        counter += 1 # 나누기 수행 횟수 +1
        if n % prime[i] == 0: # n이 prime[i]와 나누어 떨어지면
            break # for문을 탈출
    else: # 중간에 for문을 탈출하지 않았다면
        prime[ptr] = n # prime[ptr]에 n으로 값을 업데이트
        ptr += 1 # 소수가 추가됐으므로 소수의 개수도 +1

for i in range(ptr): # 0부터 ptr-1까지
    print(prime[i]) # prime 리스트의 원소를 하나씩 출력
print(f'나눗셈을 실행한 횟수: {counter}')
````

- 소수 찾기 알고리즘 개선 2
    - ex) 넓이가 100인 직사각형의 가로, 세로 변의 길이의 경우의 수(단, 각 변의 길이는 정수이고 1은 제외)  
        $2 \times 50 \\ 4 \times 25 \\ 5 \times 20 \\ 10 \times 10 \\ 20 \times 5 \\ 25 \times 4 \\ 50 \times 2$
    - 어떤 임의의 정수 n을 직사각형의 넓이로 봤을 때, 직사각형 한 변의 길이만 나눗셈을 시도하고, 그 과정에서 한 번도 나누어 떨어지지 않으면 소수로 판단
    - 소수의 조건 : **어떤 정수 n이 소수라면 n은 <u>n의 제곱근($\sqrt {n}$) 이하의 소수</u>로 나누어 떨어지지 않는다.**

````python
# 1000이하의 소수 나열하기(알고리즘 개선 2)

counter = 0 # 곱셈, 나눗셈 연산 횟수
ptr = 0 # 앞서 찾은 소수의 개수
prime = [None] * 500 # 소수를 저장할 리스트

prime[ptr] = 2 # 2는 소수
ptr += 1 # 소수의 개수 +1

prime[ptr] = 3 # 3은 소수
ptr += 1 # 소수의 개수 +1

for n in range(5, 1001, 2): # 5, 7, 9, 11, 13, 15, ..., 999까지(= 홀수)
    i = 1 # prime 리스트의 인덱스 prime[1] = 3
    while prime[i] * prime[i] <= n: # 소수의 제곱이 n보다 작거나 같은 동안 반복
        counter += 2 # 곱셈 한 번, 나눗셈 한 번 하므로 +2
        if n % prime[i] == 0: # n 나누기 prime[i]가 나누어 떨어지면
            break # while문을 탈출
        i += 1 # 나누어 떨어지지 않은 경우에만 +1
    else: # while문을 탈출하지 않았다면(= 끝까지 나누어 떨어지지 않았다면)
        prime[ptr] = n # n은 새로운 소수이므로 prime 리스트에 추가
        ptr += 1 # 소수의 개수 +1
        counter += 1 # break를 통해 while문을 탈출했다면 prime[i] * prime[i]의 곱셈은 연산 횟수에 추가되지 않으므로 나누기 연산 횟수 한 번만 추가

for i in range(ptr):
    print(prime[i])
print(f'곱셈과 나눗셈을 실행한 횟수: {counter}')
````

- 첫번째 알고리즘의 계산 횟수 78,022에서, 한 번 개선을 한 알고리즘은 14,622, 또 한 번 개선한 알고리즘은 3,774로 연산 횟수가 점점 줄었다.
- 대신 첫번째 알고리즘보다 보다 많은 메모리를 요구한다.(먼저 찾은 소수를 저장하는 리스트)

    => <u>**빠른 알고리즘은 그만큼 많은 메모리를 요구한다.**</u>

- 얕은 복사(Shallow Copy)와 깊은 복수(Deep Copy)
    - 얕은 복사 : 객체가 갖는 멤버의 값을 새로운 객체로 복사할 때 객체가 참조 자료형의 멤버를 포함하는 경우. 참조값만 복사
    - 깊은 복사 : 참조값 뿐만 아니라 참조하는 객체 자체를 복사(원본이 업데이트되도 바뀌지 않음)

````python
# 얕은 복사
x = [[1, 2, 3], [4, 5, 6]]
y = x.copy() # 얕은 복사
x[0][1] = 9
print(x)
print(y) # 복사한 y도 같이 수정됨
````

````python
# 깊은 복사
import copy

x = [[1, 2, 3], [4, 5, 6]]
y = copy.deepcopy(x) # 깊은 복사
x[0][1] = 9
print(x)
print(y) # 복사된 y는 수정되지 않음
````
